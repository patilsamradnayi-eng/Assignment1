```python
#Demonstration of Python Data Types

# Integer
a = 10
print("Integer value:", a)

# Float
b = 5.75
print("Float value:", b)

# String
c = "Hello Python"
print("String value:", c)

# Boolean
d = True
print("Boolean value:", d)

# List
e = [1, 2, 3, 4]
print("List value:", e)

# Tuple
f = (10, 20, 30)
print("Tuple value:", f)

# Dictionary
g = {"Name": "samradnayi", "Age": 18}
print("Dictionary value:", g)

# Set
h = {1, 2, 3, 4}
print("Set value:", h)
```
